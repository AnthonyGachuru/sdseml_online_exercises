# Solution Exercise 2.1.1

sklearn.ensemble.RandomForestClassifier is a random forest classifier which is used to predict the data. The random forest is an estimator that fit a number of decision tree classifiers to imporove the predictive power. The Sklearn.ensemble.RandomForestClassifer has as mentioned in the exercise text many different hyper parameters that have certain functions. Below i will explain the four parameters `n_estimators`, `max_depth`,`max_features` and `bootstrap`. 

1. `n_estimators`

n_estimators is the number of decision trees in the forest. The defualt value is 100. A high number of n_estimators will make the estimation better. But it will also caurse the estimation to take longer. Furthermore, there will be a certain point where a higher number of estimations will not imporove the result. So if the n_estimators is too low the result will be inaccurate, but if you set the n_estimators too high the result will be accurate but it will take a very long time to estimate and a lower number of n_estimators chould maybe decrease the time, without significantly hurting the results. 


2. `max_depth` 

max_depth is a parameter controlling the max depth of the trees (the depth of the tree is the length of the longest path from root to leaf in the decision trees). So by setting a max depth it is possible to control and limmit the time, so the estimation is stopping at a certain point which can be preferable if the tree is very large. The default value is none, which means that the nodes in the decision tree will expand until the leaves are pure, hence all leaves are exploited from root to leaf. Setting the max_depth to none can give problems since it in some cases can take a lot of computer power and thereby time to estimate. On the other hand a high value of max_depth will make the results insignificant. 


3. `max_features`

max_features is determing the number of features to concider when looking for a split. (The branches in the tree represents the feauteres that leads to to the leaves which represents the the class labels). The max_features can be set to be calculated as a function of the number of features (n_features) in the tree or just a number. The different ways to set max_features is auto, which is the default, float, int, sqrt, log2 or None. 

* If max_features=auto the maximum number of features the classifier will consider when looking for a split will be calculated as "max_features=sqrt(n_features)"

* If max_features=int the maximum number of features the classifier will consider when looking for a split will be the maximum features. 

* If max_features=float the maximum number of features the classifier will consider when looking for a split will be calculated as "int(max_features * n_features)"

* If max_features=sqrt the maximum number of features the classifier will consider when looking for a split will be calculated as "max_features=sqrt(n_features)"

* If max_features=log2 the maximum number of features the classifier will consider when looking for a split will be calculated as "max_features=log2(n_features)"

* If max_features=None the maximum number of features the classifier will consider when looking for a split will be calculated as "max_features=n_features"

If max_feaures is set at None, it will consider all features which will give good result in regression problems, since all features is taking into account, but in some cases this will be time consuming. For classification problems it can be optimal to only include a subsample of the features when looking for a split.  


4. `bootstrap` 


Bootstrap is a boolean value which is either True or False and is determine whether boostrap values is used to build the decision tree. The default value is True. (A boostrap sample is a samaller sample that has been resampled from a large orginal sample. Often a lot of smaller samples as been bostrapped form the orginal sample). If the value is `True` the tree was build from a boostrap sample of the original data, if `False` the whole dataset has been used to build the decision tree. 

Boostrapping is a good way to check the stability of the results and the boostrap samples are used as a default in a random forrest. There can be some problems by using a boostrap sample if the sample size is very small.

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits

# Get wine data
data_wine = load_wine()
X_wine = data_wine['data']
y_wine = data_wine['target']

# Get digits data
data_digits = load_digits()
X_digits = data_digits['data']
y_digits = data_digits['target']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#PCA plot. 

pca = PCA()


#Wine data
X_wine_std = StandardScaler().fit_transform(X_wine)
X_wine_pca = pca.fit_transform(X_wine_std)[:, :2]

#Digits data
X_digits_std = StandardScaler().fit_transform(X_digits)
X_digits_pca = pca.fit_transform(X_digits)[:, :2]
```

> **Ex. 4.2.1:** What does it mean that a tree is _honest?_ In particular what are the implications in terms of 
> * The intuition for why honesty is required in order to get good local treatment effect estimates?
> * The practical implementation of the DT algorithm?

# Solution for exercise 4.2.1

When is a tree honest?
In the paper "recursive partitioning for heterogeneous causal effects" by Susan Athey and Guido Imbens the term honest about a model if the model does not use the same information for selecting the structure and estimation given the model structure. The optain the honest tree, the data must be split into two training data sets, where one is used for building the decision trees and the other is used for estimating the treatment effects within the leaves of the tree. So according to Athey and Imbens a honest tree is when the sample is splitted into to and deffirent training samples is used for estimation and structuring the decision tree.   

Why is honesty required in order to get good local treatment effect estimates?
One of the advantages of using a honest tree is that there is no retrictions on the model complexity. The fact that we have splitted the data does that it is possible to focus on the variance instead of the bias-variance which would accure when not using a honest tree. The honesty is also a good way to insure that the number of observations in the leafs is high, thereby we will minimize the effect from of small leafs. This is a way to insure good local tratment effect estimates. 

What is the practical implementation of the DT algorithm? 
The practical implementation of the the decision tree algorithm is to make complex and large datasets into something understanderble, which we can get a lot of information from. The DT algorithm does not need much data preperation. Whit the honest tree approach the hetrogenity is ensured.
