```python
from scipy import integrate
import numpy as np
from numpy import log,linspace
import matplotlib.pyplot as plt

def Integrate(f,a,b):
    I, _ = integrate.quad(f,a,b)
    return I

f = lambda t: 1/np.log(t)
Li = []
for i in np.linspace(10,10**10, 100):
    Li.append(Integrate(f, 2, i))
```

```python
#conda install publib
from publib import set_style, fix_style

# it doesn't seem to take the style as i intended, im not sure how im using it wrong.

set_style('origin')

x = np.linspace(10,10**10, 100)
y = Li
#plt.figure()
ax = plt.subplot()
ax.plot(x,y)
plt.xlabel('n')
plt.ylabel('Number of Primes')
plt.yscale('log')
plt.xscale('log')

fix_style('origin')

plt.show()
```
