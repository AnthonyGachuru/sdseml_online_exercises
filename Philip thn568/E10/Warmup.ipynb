```python
import numpy as np
import matplotlib.pyplot as plt
#Todays quick warmup
```

```python
# Q1: (qa)^2+(qb)^2 = (qc)^2 <=> q^2(a+b)=q^2c^2
# which holds for all q if (a,b,c) is a Pythagorean triple
```

```python
# Q2: Note that thus function returns a list in which some triples are multiples of earlier ones.
def PT(n):
    squares = [i**2 for i in range(1,n+1)]
    a,b = [],[]
    for i in range(0,n):
        for j in range(i,n):
            if squares[i]+squares[j] in squares:
                a.append(squares[i]**(1/2))
                b.append(squares[j]**(1/2))
    return np.column_stack((a,b))
```

```python
results = PT(2000)
fig, ax = plt.subplots()
ax.scatter(results[:,0],results[:,1],c="black",linewidts = 1)
```
