```python
import math
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
#No need to construct Integrate() as it is identical to integrate.quad()[0]
n  = np.arange(10,10**100,(10**100-10)/100)
pi = np.ones((1,100)).T.reshape(-1)
for i in range(0,n.shape[0]):
    pi[i] = integrate.quad(lambda t: 1/math.log(t),2,n[i])[0]
```

```python
fig, ax = plt.subplots()
ax.plot(pi,n,c="black")
ax.set_xlabel("n")
ax.set_ylabel("$\sim$#Primes<n")
ax.set_title("Warmup")
```
