```python
from scipy import integrate
import numpy as np
from numpy import log,linspace
import matplotlib.pyplot as plt

def Integrate(f,a,b):
    I, _ = integrate.quad(f,a,b)
    return I

```

```python
f = lambda t: 1/np.log(t) 


range = np.linspace(10, 10**10, 100) # 100 numbers between 10^1 and 10^10
L_in = [Integrate(f,2,i) for i in range]
   
```

```python

# Matplotlib style cheat
plt.style.use('science')

with plt.style.context(['science']):
    fig, ax = plt.subplots(figsize = (10,7))
    plt.plot(range,L_in) 
    ax.set_title('Approximation nmber of primes less than n', size=25)
    ax.set_ylabel('Number of primes below n', size=20)
    ax.set_xlabel('n', size=20)
    plt.yscale('log')
    plt.xscale('log')
```
