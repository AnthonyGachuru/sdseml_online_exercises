Description:
Nodes no longer have the same possibility of attaching to every node in the network like in the ER, now nodes have a preference (a higher probability) to attach to nodes that are already highly connected. This phenomonen is known as the "power law" and is reconized from several real-life network where so called "hubs" are created -> nodes with a lot of, and a growing ammount of, connections.

```python
n=100
m = 2
G = nx.barabasi_albert_graph(n, m, seed=None)
```

```python
nx.draw(G)
```

```python
#Create list of degrees:
d = []

for i in range(len(_G)):
    k = G.degree[i]
    d.append(k)
    i+1
    
print(d)
```

```python
plt.figure(figsize=[10,8])
n, bins, patches = plt.hist(x=d, bins=10, color='#0504aa',alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Degree',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Degree Distribution',fontsize=15)
plt.show()
```

In the ER network the degrees are somewhat normaly distributed due to the equal possibility of attachment to every other node. In the BA network a single node has 25 connections, meaning 1/4 of all nodes in the network is connected to this particular node while a lot of the other nodes has only two connections. This illustrates the power law.

Other interesting measures could have been the clustering coefficient, the average path length or the network diameter.

Now that you know how to build a network in `NetworkX`, visualize it and plot some statistics, you are ready to analyze today's dataset. We will be using [this facebook dataset](http://socialnetworks.mpi-sws.org/data-wosn2009.html) collected by the researchers at the Max Planck Institute. Go ahead and download the list of wall posts. Put the file in the same directory as this notebook and load it like:

```python
data = pd.read_csv(
    "facebook-wall.txt.anon",
    delimiter="\t",
    header=None,
    names=['user1', 'user2', 'timestamp']
)

data.head(10)
```

As you can see this is a *temporal network* (in that edges are timestamped). Time is given in UNIX timestamps. Though there are ways to model such
a network in its entire complexity, we'll start simple and just take a slice of it out and treat that as a static
network. You can create a `nx.DiGraph` from a slice of time like:

```python
t0 = data.timestamp.max() - 86400 * 7  # Lower bound of time-slice (here: 7 days before last wall post)
t1 = data.timestamp.max()              # Upper bound of time-slice

# Subset of the pandas dataframe
data_t = data.loc[data.timestamp.between(t0, t1)]

# Count the number of times each link occurs and store that in a new 'weight' column
data_t = data_t.groupby(['user1', 'user2']).size().reset_index(name='weight')

# Create a `nx.DiGraph` from this dataframe
G_t = nx.from_pandas_edgelist(data_t, 'user1', 'user2', 'weight', create_using=nx.DiGraph)
```

> **Ex: 5.2.1**: Convert the above sample code into a function where you can easily input `t0` and `t1` of the slice, and get a `nx.DiGraph` in return. Create the aggregated network for the last 365 days of the data. Print the number of nodes and edges.

```python
def facebook(t0, t1):

# Subset of the pandas dataframe
    data_t = data.loc[data.timestamp.between(t0, t1)]

# Count the number of times each link occurs and store that in a new 'weight' column
    data_t = data_t.groupby(['user1', 'user2']).size().reset_index(name='weight')

# Create a `nx.DiGraph` from this dataframe
    G_t = nx.from_pandas_edgelist(data_t, 'user1', 'user2', 'weight', create_using=nx.DiGraph)

    return G_t
```
