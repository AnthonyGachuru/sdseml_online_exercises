```python
from scipy import integrate
from numpy import log, linspace
import matplotlib.pyplot as plt
import pandas as pd  
```

```python
def Integrate(f, a, b):
    L = []
    for i in b:
        I, _ = integrate.quad(f, a, i)
        L.append([I])
    return L
```
