# Warm-Up: Pythagorean triple

```python
tuple = (3,4,5)
```

```python
def triple(tuple):
    return tuple[0]**2 + tuple[1]**2 == tuple[2]**2
```

```python
triple(tuple)
```

```python
# q1
```

```python
import random

def q1(tuple):
    if (triple(tuple)==True):
        q = random.randint(0,10000)
        qtuple = q*tuple
        
        if (triple(qtuple)==True):
            print('It is still a Pythagorean triple')
    else:
        print('It is not a Pythagorean triple')
```

```python
q1(tuple)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def tuple_plot(n):
    a_t = []
    b_t = []
    for a in range(1, n):
        for b in range(1, n):
            c = math.sqrt(a**2 + b**2)
            if (c.is_integer()==True):
                    a_t.append(a)
                    b_t.append(b)
                    continue
    plt.figure(figsize=(15,15))                
    plt.scatter(a_t, b_t, c='black', alpha=0.3, linewidths=0.3)
    plt.xlabel("a")
    plt.ylabel("b")
```

```python
tuple_plot(5000)
```
