```python
from scipy import integrate
from numpy import log, linspace
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Integrate(f, a, b):
    I, _ = integrate.quad(f, a, b)
    return I

Li = lambda t: 1/log(t)
N = np.logspace(1,10, num=100)
pi = [Integrate(Li, 2, n) for n in N]
```

```python
repo = "https://raw.githubusercontent.com/nicoguaro/matplotlib_styles/master"
style = repo + "/styles/clean.mplstyle"
with plt.style.context(style):
    plt.plot(N, pi)
    plt.xlabel('n')
    plt.ylabel('Number of Primes')
    plt.yscale('log')
    plt.xscale('log')
    plt.show()           
```
