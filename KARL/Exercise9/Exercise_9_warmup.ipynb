```python
from scipy import integrate      
from numpy import log, linspace  
import matplotlib.pyplot as plt  
```

```python
# define function
def f(x):
    return 1/log(x)

# create list with limits
limits = list(linspace(10, 10**10, 100))

# funtion for integration using integrate.quad from scipy
def Integrate(f,a,b):
    I, _ = integrate.quad(f,a,b)
    return I

# integrate and append to list
integrals = []
for n in limits:
    i = Integrate(f, 2, n)
    integrals.append(i)
```

```python
# plot with style sheet publib
from publib import set_style, fix_style
set_style('article')

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(limits, integrals)
ax.legend(['Logarithmic integral'])
ax.set(title = 'Plot of logarithmic integral',
       ylabel = r'$\int_{2}^{n}{\frac{1}{log(t)}}dt$',
       xlabel = 'n')

fix_style('article')
```
