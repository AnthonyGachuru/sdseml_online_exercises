```python
# Function for number of primes less than n:

# we have that li(x) = int^b_a dt/log(t)
# and we have that the Eulerian logarithmic integral is:
# Li(n) = li(n) - li(2)

import numpy as np
from scipy import integrate
from numpy import log, linspace
import matplotlib.pyplot as plt
import math

def Integrate(f, a, b):
    I, _ = integrate.quad(f, a, b)
    return I

n =  np.linspace(10,10**10,100)
y = lambda t : 1/(math.log(t))

plt.plot(n,[Integrate(f=y, a=2, b=i) for i in n])
plt.yscale("log")
#plt.xscale("log")
plt.show()
```
