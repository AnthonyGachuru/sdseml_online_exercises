> **Ex. 2.1.1**: The performance of a random forest classifier is highly sensitive to hyper parameters. Therefore, you should be intimately familiar with what the most important ones do. For each point below, explain what the hyper parameter pertaining to `sklearn.ensemble.RandomForestClassifier` controls, and how setting it either too low or too high (or True/False) might hurt model performance:
1. `n_estimators`: Setting the number of estimators to 1 gives you the decision tree. We only consider orthogonal splits and thus never linear combinations of x and y. Adding more trees makes the split more detailed. Thus, a many trees will perhaps overfit and less will perhaps underfit. 
2. `max_depth`: The depth of each tree - amount of tress or nodes. The lowest you can get is of course 1. The maximum value of max depth is when you have only one observation in each leave.
3. `max_features`: The maximal number of features in each node split. 
4. `bootstrap`: Whether you bootstrap or not. Set to True as default. If False the whole dataset is used to build each tree.

```python
# 3.1.2.1 Digits dataset 

#Import
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits

%matplotlib inline
import matplotlib.pylab as plt
from sklearn.preprocessing import StandardScaler

#Dimension reduction methods
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 
from sklearn.manifold import TSNE
import umap
```

```python
# Get digits data
data_digits = load_digits()
X_digits = data_digits['data']
y_digits = data_digits['target']

# PCA
X_digits_pca = PCA().fit_transform(X_digits)

# LDA
X_digits_lda = LDA().fit_transform(X_digits, y_digits)

# t-SNE 
X_digits_tsne = TSNE().fit_transform(X_digits)

# UMAP 
X_digits_umap = umap.UMAP().fit_transform(X_digits)

#Set the subplot (2 rows with 2 columns)
fig, ax = plt.subplots(2,2, figsize = (10,10))

#The first subplot  
ax[0,0].scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_digits, label = ' ')

ax[0,0].set_xlabel('1st component', fontsize = 10) 
ax[0,0].set_ylabel('2nd component', fontsize = 10) 
ax[0,0].set_title('PCA')
ax[0,0].grid(True)

#The second subplot 
ax[0,1].scatter(X_digits_lda[:, 0], X_digits_lda[:, 1], c=y_digits, label = 'LDA')

ax[0,1].set_xlabel('1st component', fontsize = 10) 
ax[0,1].set_ylabel('2nd component', fontsize = 10) 
ax[0,1].set_title('LDA')
ax[0,1].grid(True)

#The third subplot 
ax[1,0].scatter(X_digits_tsne[:, 0], X_digits_tsne[:, 1], c=y_digits, label = 't-SNE')

ax[1,0].set_xlabel('1st component', fontsize = 10) 
ax[1,0].set_ylabel('2nd component', fontsize = 10)
ax[1,0].set_title('t-SNE')
ax[1,0].grid(True)

#The fourth subplot 
ax[1,1].scatter(X_digits_umap[:, 0], X_digits_umap[:, 1], c=y_digits, label = 'UMAP')

ax[1,1].set_xlabel('1st component', fontsize = 10) 
ax[1,1].set_ylabel('2nd component', fontsize = 10) 
ax[1,1].set_title('UMAP')
ax[1,1].grid(True)
```

```python
# Wine data
data_wine = load_wine()
X_wine = data_wine['data']
y_wine = data_wine['target']

# We want to standardize the wine dataset so it looks like it is standard normally distributed with mean 0 and
# unit variance.
# This is done as our dimensionality reduction methods assume that our data reminds of something that is 
# standard normally distributed. 
# The reason why we do not have standardize the digits dataset is because it already is "standardized". 
# For instance, different from the wine dataset all features are already measured in the same units.  

# Standardize
sc = StandardScaler()
X_std = sc.fit_transform(X_wine)

#PCA
X_wine_pca = PCA().fit_transform(X_std)

#LDA 
X_wine_lda = LDA().fit_transform(X_std, y_wine)

#t-SNE 
X_wine_tsne = TSNE().fit_transform(X_std)

#UMAP 
X_wine_umap = umap.UMAP().fit_transform(X_std)

#Set the subplot (2 rows with 2 columns)
fig, ax = plt.subplots(2,2, figsize = (10,10))

#The first subplot  
ax[0,0].scatter(X_wine_pca[:, 0], X_wine_pca[:, 1], c=y_wine, label = ' ')

ax[0,0].set_xlabel('1st component', fontsize = 10) 
ax[0,0].set_ylabel('2nd component', fontsize = 10) 
ax[0,0].set_title('PCA')
#ax[0,0].legend()
ax[0,0].grid(True)

#The second subplot 
ax[0,1].scatter(X_wine_lda[:, 0], X_wine_lda[:, 1], c=y_wine, label = 'LDA')

ax[0,1].set_xlabel('1st component', fontsize = 10) 
ax[0,1].set_ylabel('2nd component', fontsize = 10) 
ax[0,1].set_title('LDA')
ax[0,1].grid(True)

#The third subplot 
ax[1,0].scatter(X_wine_tsne[:, 0], X_wine_tsne[:, 1], c=y_wine, label = 't-SNE')

ax[1,0].set_xlabel('1st component', fontsize = 10) 
ax[1,0].set_ylabel('2nd component', fontsize = 10)
ax[1,0].set_title('t-SNE')
ax[1,0].grid(True)

#The fourth subplot 
ax[1,1].scatter(X_wine_umap[:, 0], X_wine_umap[:, 1], c=y_wine, label = 'UMAP')

ax[1,1].set_xlabel('1st component', fontsize = 10) 
ax[1,1].set_ylabel('2nd component', fontsize = 10) 
ax[1,1].set_title('UMAP')
ax[1,1].grid(True)
```

**Ex 3.1.2.3 The differences and applicability of the different methods for different data**
> The main difference between the methods is the trade-off between the interpretation and variance of the data. For instance, you can use UMAP and reduce the variance without being able to intepret the features of the data. Thus, as we discussed in class, PCA is often preferred over the other methods as you are actually able to interpret the data. 
>
> When examing the methods on the two datasets it is obvious that they behave differently dependent on whether the it is standardized or not. It is evident that the difference between the methods is smaller on a standardized dataset.

**Ex 4.2.1 Honest trees**
> A model is honest when no restrictions are placed on the complexity of the model. Practically, that is accomplished by splitting the training data into a part for constructing the tree and another part for estimation. For instance, the same information must not be used for selecting the model structure as for estimating the data. 
> 
> An honest estimation outperforms adaptive estimation, and thus gives good local treatment effect estimates, because the honest approach eliminates bias and optimize the test results.
