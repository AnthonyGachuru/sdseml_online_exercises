```python
from scipy import integrate
from numpy import log, linspace
import matplotlib.pyplot as plt

results = []
def Integrate(f,a,b): 
    for i in b:
        I, _ = integrate.quad(f,a,i)
        results.append(I)
    return
```

```python
f = lambda x: 1/log(x)
b = linspace(10,10**10,num=100)
a = 2
Integrate(f,a,b)
```

```python
print(results)
```
