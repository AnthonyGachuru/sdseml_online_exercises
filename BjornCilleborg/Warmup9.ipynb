```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

def Integrate(f, a, b):
    I, _ = integrate.quad(f, a, b)
    return I
```

In this warm up we seek to write an logarithmic integration function, $\pi(n)$, that approximate the number of prime numbers less than n.

```python
# We first define our linear space of n

n = np.linspace(10, 10**100, 100) # start, end, buckets

# Create the lambda function that we want to integrate over
f = lambda x : np.log(x)**(-1)

# Integrate over linsapce
pi_n = [Integrate(f,2,i) for i in np.linspace(10,10**10,100)] 
```

```python
# Plotting

fig, ax = plt.subplots()
ax.plot(n, pi_n)

ax.set_xlabel("n", fontsize = 12)
ax.set_ylabel(" # of primes less than n", fontsize = 12)
ax.set_title("Approximation of primes function, $\pi(n)$",fontsize=14)
plt.style.use('fivethirtyeight')
plt.show()
```
