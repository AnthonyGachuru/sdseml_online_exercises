### Q2: Code to find all pythagorean triples
*I ended up not using this approach since it was much faster to make a quick and dirty solution*

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def gcd(a, b):
    """
    This function returns the greatest common divisor (gcd) of two integers
    
    input
    a: int
    b: int
    
    return
    a: int
    - the gcd of a and b
    """
    while b != 0:
        temp_a = a
        a = b
        b = temp_a % b    
        
    return a


def create_pythagorean_triples(max_a=10000,max_b=10000):
    pythagorean_triples = []
    list1 = list(range(1000))
    list2 = list(range(1000))

    # Create a list of tuples defining pairs of coprime numbers
    coprimes     = [(m, n) for m in list1 for n in list2 if gcd(m, n) == 1]

    # keep coprimes without both m and n being odd
    not_both_odd = [(m, n) for m,n in coprimes if ((m%2==0) | (n%2==0))] 

    for m,n in not_both_odd:
        a = m**2 - n**2
        b = 2*m*n
        c = m**2 + n**2
        if np.min([a,b,c])>0:
            if a<=max_a:
                if b<=max_b:
                    pythagorean_triples.append([a,b,c])

    return pythagorean_triples


```

### Q2: Quick and dirty solution
*This method simply evaluates a meshgrid of a and b values and checks whether $\sqrt{a^2+b^2}$ is an integer - which implies (a,b,c) is a pythagorean triple.*

```python
max_a = 10000
max_b = 10000

x_cord = np.arange(max_a)
y_cord = np.arange(max_b)

xx,yy           = np.meshgrid(x_cord,y_cord)
c               = np.sqrt(xx**2 + yy**2)
c_final         = c.copy()
c_final[c%1!=0] = 0
c_final[c%1==0] = 1

plt.contour(x_cord,y_cord,c_final)
plt.xlabel('a value')
plt.ylabel('b value')
plt.title('Visualization of pythagorean triples')
```
