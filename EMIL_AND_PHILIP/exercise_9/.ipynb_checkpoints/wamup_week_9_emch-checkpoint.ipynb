```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from publib import set_style, fix_style

## Load the latex custom style sheet using publib
set_style(['latex'])
```

```python
def Integrate(f,a,b):
    """
    Computes the definite integral of a function f. 
    Integrate func from a to b (possibly infinite interval) using a technique from the Fortran library QUADPACK.

    arguments:
    f: function
    - should return a float or int
    
    a: int
    - lower limit of the integral
    
    b: int
    - upper limit of the integral
    
    returns:
    numeric_result: float
    - the numerical approximation of the integral

    """
    
    numeric_result, numeric_err = integrate.quad(f, a, b)
    return numeric_result


## Define the logarithmic integral
def logarithmic_integral(n):
    """
    Computes the approximation of pi(n) [the number of primes less than n] using scipy.integrate.quad()
    
    arguments:
    n: int
    - upper integral limit
    
    returns:
    Number of primes less than n (float)
    
    """
    f = lambda t:1/np.log10(t)
    return Integrate(f,2,n)
```

```python
## Calculate x and y values
n      = np.linspace(10,10**10,100)
li_n   = [logarithmic_integral(n) for n in n]

## Create plot
fig, ax = plt.subplots()
ax.plot(n,li_n)

## Set aesthetics
ax.set_xlabel("n")
ax.set_ylabel("Number of primes less than n",fontsize=16)
ax.set_title("Approximation of $\pi(n)$")
```
