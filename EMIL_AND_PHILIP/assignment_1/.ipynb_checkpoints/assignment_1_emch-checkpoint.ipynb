## Problems from exercise set 2
> **Ex. 2.1.1**: The performance of a random forest classifier is highly sensitive to hyper parameters. Therefore, you should be intimately familiar with what the most important ones do. For each point below, explain what the hyper parameter pertaining to `sklearn.ensemble.RandomForestClassifier` controls, and how setting it either too low or too high (or True/False) might hurt model performance:
1. `n_estimators`
2. `max_depth`
3. `max_features`
4. `bootstrap`

##### ANSWER TO ex 2.1.1.
* *n_estimators is the number of random decision trees generated. Normally, a larger number a trees increases both the train and test scores. However, it is computionally heavy to estimate a lot of trees. Moreoever, there is a diminishing return to score performance. Hence, in practice one usually use something between 100-1000 n_estimators depending on the data size and computer speed.*

* *max_depth is the maximum 'depth' of each generated tree. That is, the number of allowed partitions inside the tree. If the max_depth is set to unlimited each decision tree only contains pure splits. This obviously leads to strong overfitting. Hence, in practise the max_depth should be set well below unlimited to aviod overfitting.*

* *max_features is the number of features to consider for every split during a generation of a random decision tree. Naturally, it takes longer time if the computer has to consider all available features. However, it can also improve the test score precision if the computer is allowed to use all features. In my experience this is a hyperparameter which can be worth setting to 1 (aka considers all features). Obviously, one should conduct gridsearch to investigate whether this is a good apprach.*

* *bootstrap determines whether each generated decision tree should be created based on a bootstraped sample of the full data set. This is a main property of random forest classifiers. To leverage on the idea of comparing multiple decision trees to make a prediction, we want the decision trees to be independent of each other. By bootstrapping we increase the independence of each decision tree. For this reason the default value of bootstrap is True.*

> **Ex. 2.1.2**: For `n_estimators > 1`how should one set the hyper-parameters `max_features` and `bootstrap` so that all the trees in the ensemble end up identical?

##### ANSWER TO ex 2.1.2.
*If n_estimators is above 1 we create more than one random decision tree. If we want all trees to be identical we could set max_features to 1 (as a float) and bootstrap to False. This would imply that each decision tree is generated from identical data. Hence, the decision trees would be identical. This would not be a good approach if we want to reduce overfitting.*

```python
import matplotlib.pyplot as plt
import umap
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

## Set control options
n_components = 2 ## Number of components

# Get wine data
data_wine = load_wine()
X_wine = data_wine['data']
y_wine = data_wine['target']

# Get digits data
data_digits = load_digits()
X_digits = data_digits['data']
y_digits = data_digits['target']

## Standardize wine data set
scaler       = StandardScaler()
X_wine_std   = StandardScaler().fit_transform(X_wine)

## PCA
pca_wine     = PCA(n_components=n_components).fit_transform(X_wine)
pca_wine_std = PCA(n_components=n_components).fit_transform(X_wine_std)
pca_digits   = PCA(n_components=n_components).fit_transform(X_digits)

## LDA 
lda_wine     = LinearDiscriminantAnalysis(n_components=n_components).fit_transform(X_wine,y_wine)
lda_wine_std = LinearDiscriminantAnalysis(n_components=n_components).fit_transform(X_wine_std,y_wine)
lda_digits   = LinearDiscriminantAnalysis(n_components=n_components).fit_transform(X_digits,y_digits)

## TSNE
tsne_wine     = TSNE(n_components=n_components).fit_transform(X_wine,y_wine)
tsne_wine_std = TSNE(n_components=n_components).fit_transform(X_wine_std,y_wine)
tsne_digits   = TSNE(n_components=n_components).fit_transform(X_digits,y_digits)

## UMAP
umap_wine     = umap.UMAP(n_components=n_components).fit_transform(X_wine,y_wine)
umap_wine_std = umap.UMAP(n_components=n_components).fit_transform(X_wine_std,y_wine)
umap_digits   = umap.UMAP(n_components=n_components).fit_transform(X_digits,y_digits)
# Get wine data
data_wine = load_wine()
X_wine = data_wine['data']
y_wine = data_wine['target']

# Get digits data
data_digits = load_digits()
X_digits = data_digits['data']
y_digits = data_digits['target']
```

```python
wine_loop = [pca_wine,lda_wine,tsne_wine,umap_wine]

fig,ax = plt.subplots(2,2)
fig.suptitle('DIM. REDUCTION\nOF WINE DATA SET (NON-STANDARDIZED)', fontsize=14,y=1.05)
fig.tight_layout(pad=3.0)
for i,transformed_data in enumerate(wine_loop):
    df_plot = pd.DataFrame(data=transformed_data,columns=['component_1','component_2']).assign(class_type = y_wine)
    classes = df_plot.class_type.unique()
    
    for class_type in classes:
        df_plot_tmp = df_plot.query(f'class_type=={class_type}')
        if(i==0):
            ax[0,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,0].set_title('PCA')
        if(i==1):
            ax[0,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,1].set_title('LDA')
        if(i==2):
            ax[1,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,0].set_title('T-SNE')
        if(i==3):
            ax[1,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,1].set_title('UMAP')
```

```python
wine_std_loop = [pca_wine_std,lda_wine_std,tsne_wine_std,umap_wine_std]

fig,ax = plt.subplots(2,2)
fig.suptitle('DIM. REDUCTION\nOF WINE DATA SET (STANDARDIZED)', fontsize=16,y=1.05)
fig.tight_layout(pad=3.0)
for i,transformed_data in enumerate(wine_std_loop):
    df_plot = pd.DataFrame(data=transformed_data,columns=['component_1','component_2']).assign(class_type = y_wine)
    classes = df_plot.class_type.unique()
    
    for class_type in classes:
        df_plot_tmp = df_plot.query(f'class_type=={class_type}')
        if(i==0):
            ax[0,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,0].set_title('PCA')
        if(i==1):
            ax[0,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,1].set_title('LDA')
        if(i==2):
            ax[1,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,0].set_title('T-SNE')
        if(i==3):
            ax[1,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,1].set_title('UMAP')
```

```python
digit_loop = [pca_digits,lda_digits,tsne_digits,umap_digits]

fig,ax = plt.subplots(2,2)
fig.suptitle('DIM. REDUCTION OF DIGIT DATA SET', fontsize=16, y = 1.05)
fig.tight_layout(pad=3.0)
for i,transformed_data in enumerate(digit_loop):
    df_plot = pd.DataFrame(data=transformed_data,columns=['component_1','component_2']).assign(class_type = y_digits)
    classes = df_plot.class_type.unique()
    
    for class_type in classes:
        df_plot_tmp = df_plot.query(f'class_type=={class_type}')
        if(i==0):
            ax[0,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,0].set_title('PCA')
        if(i==1):
            ax[0,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[0,1].set_title('LDA')
        if(i==2):
            ax[1,0].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,0].set_title('T-SNE')
        if(i==3):
            ax[1,1].scatter(df_plot_tmp.component_1,df_plot_tmp.component_2)
            ax[1,1].set_title('UMAP')
```

### COMMENT ON DIMENSIONALITY PLOTS

* Problem with non-standardized wine data: The wine data contains features with very different scaling and levels. Since the methods are sensitive to scaling and levels of the features, they perform poorly when we do not standardize our data prior to the component analyses. The problem seems to be most severe for the non-linear methods (T-SNE and UMPA) which are a lot better after standardization. The features in the digit data set are similar in scale and levels. Hence, standardization is less important for this data set.
* Differences across methods: Generally, the models which use class information in the component analysis (LDA, T-SNE and UMAP) are visually better at separating the classes using the first two components. Moreover, the non-linear methods (T-SNE and UMAP) are better at seperating the classes. This is especially true for the digits data set, which is highly non-linear.

In this problem we will try to implement and understand some of the ideas used in [Athey, Imbens (2015)](https://www.pnas.org/content/pnas/113/27/7353.full.pdf) to develop _Honest Inference_ in desicion tree models. The paper begins by covering honesty in a setting of population averages, and for estimating conditional means; so you will need to look towards the second half of the paper to get an impression of it's use for treatment-effect estimation.

> **Ex. 4.2.1:** What does it mean that a tree is _honest?_ In particular what are the implications in terms of 
> * The intuition for why honesty is required in order to get good local treatment effect estimates?
> * The practical implementation of the DT algorithm?

*Answer to ex 4.2.1:* The honesty refers to the fact that the methods seek to avoid data leakages across the classfication of decision trees and the estimation of treatment effect. The decision tree process essentially partitioning on largest impurity in treatment effect among the individuals in the node. Thus, if we naively estimate the local treatment effect on the same individuals as we use to create the decision trees we would expect to get biased treatment effects by construction. For this reason, it is a better approach to set aside a set of untouched individuals for which we estimate the local treatment effect AFTER we have created the decision trees. Given this set is a random set (indepedent of the training set used for the decision tree classification) the biased converges to zero. It would probably be a good idea to use a CV for this matter.
