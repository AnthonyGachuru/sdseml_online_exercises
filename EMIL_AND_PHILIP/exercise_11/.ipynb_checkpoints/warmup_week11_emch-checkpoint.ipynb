```python
fibonnaci_cache = {}
def fibonacci(n):
    
    if n in fibonnaci_cache:
        return fibonnaci_cache[n]
    
    if n==1:
        value =  1
    elif n==2:
        value =  1
    else:
        value =  fibonacci(n-1) + fibonacci(n-2)
    
    fibonnaci_cache[n] = value
    return value

#[print(fibonacci(n)) for n in range(1,50)]
```

```python
from functools import lru_cache
```

```python
import numpy as np
from functools import lru_cache


@lru_cache(maxsize=1000)
def chebyshevs_poly(n,x):
    if n==1:
        return 1
    elif n==2:
        return x
    else:
        return 2*x*chebyshevs_poly(n=n-1,x=x)-chebyshevs_poly(n-2,x=x)
    
```

```python
[chebyshevs_poly(n=5,x=x_tmp) for x_tmp in np.linspace(-1,1,10000)]
```
