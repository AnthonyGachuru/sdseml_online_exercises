```python
from scipy import integrate
from numpy import log, linspace
import matplotlib.pyplot as plt
import matplotlib.ticker

def Integrate(f,a,b):
    I, _ = integrate.quad(f,a,b)
    return I

func = lambda x: 1/log(x)

n_vals = linspace(10,10**10,100)

approx = [Integrate(func,2,i) for i in linspace(10,10**10,100)]

plt.style.use('bmh')

fig, ax = plt.subplots(figsize=(12,8))
plt.plot(linspace(10,10**10,100), approx,marker='o', markersize=5, label='Li(n)',linestyle="None")
plt.ylabel("Li(n)")
plt.xlabel("n")
ax.ticklabel_format(useOffset=False, style='plain')
ax.get_xaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.legend()
plt.show()
```
