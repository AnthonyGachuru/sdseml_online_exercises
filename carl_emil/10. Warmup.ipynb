### Q.1: Show that if $(a, b, c)$ is a Pythagorean triple, so is $(qa, qb, qc)$ for $q\in \mathbb{N}$

> \begin{align}
  (qa)^2 + (qb)^2 &= (qc)^2 \\
  q^2(a^2 + b^2) &= q^2 c^2 \\
  q^2 a^2 + q^2 b^2 &= q^2 c^2 
  \end{align}
  for any $q\in \mathbb{N}$.

### Q.2: Plot in $(a, b)$ space all Pythagorean triples with $a, b < 10,000$. 

```python
import matplotlib.pyplot as plt
import numpy as np

triples = []

for a in range(1, 10000):
    for b in range(1, 10000):
        c = np.sqrt(a*a + b*b)
        if a*a + b*b == c*c:
            if c.is_integer():
                triples.append([a, b])

plt.scatter([i for i, j in triples], [j for i, j in triples], s = 0.05)
plt.plot(figsize = (14,8))
plt.xlabel('a')
plt.ylabel('b')
plt.title('Pythagorean triples for $a, b < 10,000$')
plt.show()
```
